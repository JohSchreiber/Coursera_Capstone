```python
# create map of New York using latitude and longitude values
map_datascience = folium.Map(location=[latitude, longitude], zoom_start=8)

# add markers to map
for lat, lng, company, joblocation in zip(df['latitude'], df['longitude'], df['company'], df['joblocation']):
    label = '{}, {}'.format(company, joblocation)
    label = folium.Popup(label)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7).add_to(map_datascience)  
    
map_datascience
```
